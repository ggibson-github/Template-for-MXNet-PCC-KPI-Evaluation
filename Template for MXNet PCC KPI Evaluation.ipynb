The equation below represents the L2 Normalization of an array (vector) of numbers. The output is another array of numbers that are ready to be put through matrix multiplication and dot product against another normalized array of numbers (stock closing prices in this case). This equation and the discussion about its use for this way of calculating PCC can be found in a paper online called "GPU-PCC: A GPU Based Technique to Compute Pairwise Pearsonâ€™s Correlation Coefficients for Big fMRI Data".

```python
import numpy as np
import pandas as pd
import mxnet as mx

# Here a pandas DataFrame is being created in stock and stock2 that holds all the data as a table (DataFrame)
stock = pd.read_csv("E:\\job4\\data\\GOOG.csv", delimiter=",", quotechar='"',usecols=["timestamp","close"])
stock2 = pd.read_csv("E:\\job4\\data\\GOOGL.csv", delimiter=",", quotechar='"',usecols=["timestamp","close"])

# The first row of data we want read in
startTime = "2019-06-17"
# Field names from the csv files
keyColumnName="timestamp"
# Will only be comparing the closing prices
valueColumnName="close"

start = 0
# Loop until we have found the first row that matches our startTime and store index in start
for index, row in stock.iterrows():
    start=index
    if row[keyColumnName].strip() == startTime:
        break

# Slice out the part of the array of values we want to keep (first 100 from the starting date)
stock = stock[start:start+100]
# Turn the data into a numpy array
stock = np.array(stock["close"].tolist())


start = 0
# Loop until we have found the first row that matches our startTime and store index in start
for index, row in stock2.iterrows():
    start=index
    if row[keyColumnName].strip() == startTime:
        break

# Slice out the part of the array of values we want to keep (first 100 from the starting date)        
stock2 = stock2[start:start+100]
# Turn the data into a numpy array
stock2 = np.array(stock2["close"].tolist())

# Use the numpy correlation coefficient function with the same data to verify the MXNet version
verify = np.corrcoef(stock,stock2)
print("numpy.corrcoef:")
print(verify)

# Copy the stock data onto the GPU as an array containing an array of 1D (1 column/variate) samples
stock = mx.nd.array([stock], mx.gpu())
stock2 = mx.nd.array([stock2], mx.gpu())

# Debugging info
#print(type(stock))
#print(stock.shape)
# Uncomment to see the data read in
#print(stock)

# A Graph is just a workflow template that will execute later when data is fed in
print("")
print("Build MXNet Symbol Graph >>>>>>>>>>>>>>>>")

# Define workflow variables
gpu_stock = mx.sym.Variable('gpu_stock')
mean = mx.sym.Variable('mean')
mean_array = mx.sym.Variable('mean_array')
variance = mx.sym.Variable('variance')
L2Norm = mx.sym.Variable('L2Norm')
normalizedStock = mx.sym.Variable('normalizedStock')

gpu_stock2 = mx.sym.Variable('gpu_stock2')
mean2 = mx.sym.Variable('mean2')
mean_array2 = mx.sym.Variable('mean_array2')
variance2 = mx.sym.Variable('variance2')
L2Norm2 = mx.sym.Variable('L2Norm2')
normalizedStock2 = mx.sym.Variable('normalizedStock2')

corr_matrix = mx.sym.Variable('corr_matrix')

# Not going to use this third transformation matrix so just give it a generic name as a placeholder
C = mx.sym.Variable('C')

# Setup a 1 by 1 matrix to hold the PCC value calculated
# We are only comparing two KPIs (two arrays aka vectors) for one PCC score 
# We can be expanded this for multi-dimension/variate/column evaluation
corr_matrix = mx.sym.zeros((1,1))

# Because math equations can be manipulated to evaluate in different orders and still give the same answer
# We are using a specific version of the Pearson Coefficient Correlation equation
# which allows us to use MXNet functions for L2 Normalization of all the values and 
# that allows matrix*matrix multiplication and dot product (addition of all the multiplied matrix slots)
# The reason why we want this is because these operations can be computed in parallel 
# on a GPU much faster than using one of the other forms of the PCC equation

# Calc the mean across all the stock closing prices for the 100 days
mean = gpu_stock.mean()
# Make an array (i.e. vector or 1D matrix) that is the same shape as the stock array 
# and that contains only duplicates of the mean value in each slot 
# (each slot of this array will be subtracted from an array slot in the stock array)
mean_array = mx.sym.reshape(mx.sym.repeat(mean, 100), shape=(1,100))

# This looks like simple subtraction, however it is actually an array minus an array
# This results in another array where each slot contains the result of same slots in the
# other two arrays (gpu_stock - mean_array) being subtracted

# By subtracting the mean what is left is just the amount the closing price varies from day to day
variance = gpu_stock - mean_array

# See notes and reference on what the L2 Normalization equation is about
normalizedStock = mx.symbol.L2Normalization(variance)

# Do the same for the second stock
mean2 = gpu_stock2.mean()
mean_array2 = mx.sym.reshape(mx.sym.repeat(mean2, 100), shape=(1,100))
variance2 = gpu_stock2 - mean_array2
normalizedStock2 = mx.symbol.L2Normalization(variance2)

# Now all that is left is to do a 1D x 1D matrix multiply and dot product after transposing 
# one of the arrays to act as the top row in the matrix to be multiplied
corr_matrix = mx.symbol.linalg_gemm(normalizedStock, normalizedStock2, C, transpose_b=True)

# Now the graph is fully defined and we feed in the stock data and instruct MXNet to use the GPU
workflow = (corr_matrix).bind(mx.gpu(), {'gpu_stock':stock, 'gpu_stock2':stock2, 'C':mx.nd.zeros((1,1), mx.gpu())})

# Execute the graph workflow to get the final PCC score
results = workflow.forward()

print("")
print("MXNet PCC results:")
print(results)
```
